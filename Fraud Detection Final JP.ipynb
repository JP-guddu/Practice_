# Clipping the Outliers

```python
def clip_outliers_iqr(data_column, lower_bound=0.25, upper_bound=0.75):
    q1 = np.percentile(data_column, 100 * lower_bound)
    q3 = np.percentile(data_column, 100 * upper_bound)
    iqr = q3 - q1
    
    lower_threshold = q1 - 1.5 * iqr
    upper_threshold = q3 + 1.5 * iqr   
    
    data_column_clipped = np.clip(data_column, lower_threshold, upper_threshold)
    return data_column_clipped
```

```python
df['amount'] = clip_outliers_iqr(df['amount'])
df['oldbalanceDest'] = clip_outliers_iqr(df['oldbalanceDest'])
df['newbalanceDest'] = clip_outliers_iqr(df['newbalanceDest'])
```

I chose the Random Forest model due to the highly unbalanced nature of the dataset, where the ratio of Legitimate transactions to Fraudulent transactions is approximately 99.87 to 0.13. 
This imbalance can lead to challenges in accurately classifying the minority class (Fraud). Random Forest, a model that constructs multiple decision trees, is particularly well-suited 
for such scenarios.

The ensemble nature of Random Forest, where multiple trees are built and combined, enables the model to address the class imbalance issue. By aggregating the outputs of multiple trees,
the model can effectively capture patterns in both majority and minority classes. Despite being computationally intensive, Random Forest capability to learn from various perspectives contributes
to its effectiveness in handling complex datasets like this one.

Furthermore, Random Forest decision trees operate in a hierarchical, binary decision-making manner, which aligns well with the boolean nature of our data. 
This simplifies the model understanding of intricate patterns within the features and aids in making accurate predictions.

In essence, Random Forest ability to mitigate class imbalance, interpret complex relationships, and leverage the advantages of decision trees makes it a strategic choice for achieving
a robust and accurate predictive model for our unbalanced dataset.
